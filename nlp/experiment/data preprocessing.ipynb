```python
# 데이터셋 불러오기
import pandas as pd
data1 = pd.read_excel('/content/drive/MyDrive/데캡디/감성대화/Training/train.xlsx')
data2 = pd.read_csv('/content/drive/MyDrive/데캡디/감성대화/Training/ai_hub_조하늘.csv')
```

```python
data1.head()
```

```python
data2.head()
```

```python
data1 = data1[['감정_대분류', '사람문장1']]
```

```python
data2 = data2[['content', 'label']]
```

```python
data1['감정_대분류'].value_counts()
```

```python
data2['label'].value_counts()
```

```python
data1 = data1[['사람문장1','감정_대분류']]
```

```python
data1 = data1.rename(columns={'사람문장1':'content', '감정_대분류':'label'})
```

```python
data1_happy = data1[data1['label'] == '기쁨']
data2_happy = data2[data2['label'] == '행복']
```

```python
data_happy = pd.concat([data1_happy, data2_happy])
```

```python
data1_surprised = data1[(data1['label'] == '불안') | (data1['label'] == '분노') | (data1['label'] == '당황')]
data2_surprised = data2[(data2['label'] == '놀람') | (data2['label'] == '분노')]
```

```python
data_surprised = pd.concat([data1_surprised, data2_surprised])
```

```python
data1_sad = data1[(data1['label'] == '슬픔') | (data1['label'] == '상처')]
data2_sad = data2[(data2['label'] == '슬픔')]
```

```python
data_sad = pd.concat([data1_sad, data2_sad])
```

```python
data2_neutral = data2[data2['label'] == '중립']
```

```python
data2_neutral
```

```python
data_happy['label'] = 0
data_surprised['label'] = 1
data_sad['label'] = 2
data2_neutral['label'] = 3
```

```python
df = pd.concat([data_happy, data_surprised, data_sad, data2_neutral])
```

```python
df['label'].value_counts()
```

```python
df = df.sample(frac=1)
```

```python
df.to_csv("/content/drive/MyDrive/데캡디/감성대화/sentiment_label4.csv", index=False)
```
