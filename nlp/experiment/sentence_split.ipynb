```python
from kss import split_sentences

s = "오늘 이별을 할 것 같은 느낌이 있었다. 아침에 애인을 만났는데 표정이 좋지 않았었기 때문이다. 불길한 예감은 역시나 이별을 통보 받았고, 난 헤어졌다. 지금 미치도록 괴롭다."
for sent in split_sentences(s):
    print(sent)
```
